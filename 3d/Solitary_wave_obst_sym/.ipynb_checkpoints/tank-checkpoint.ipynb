```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
#set_matplotlib_formats('svg')
import numpy as np
import matplotlib as mpl
from matplotlib  import pyplot as plt
```

```python
###pressure
N=3 # max 13
dtype=tuple(['d']*(N+1))
names=['t']
for i in range(N):
    field='p'+str(i+1)
    names.append(field)
#print dtype
#print names
gauges = np.genfromtxt(fname="pointGauge_pressure.csv",dtype=dtype,
                       delimiter=',',skip_header=1,names=names)
mpl.rcParams['figure.figsize'] = "10,"+str(N*2.5)
fig,ax = plt.subplots(N,1)
for j in range(N):
    ax[j].plot(gauges["t"][3:],gauges[names[j+1]][3:])
    #ax[j].plot(gauges["t"],gauges[names[j+1]])
```

```python
### elevation
N=2 #max 12
field='h'
dtype=tuple(['d']*(N+1))
names=['t']
for i in range(N):
    names.append(field+str(i+1))
#print dtype
#print names
gauges = np.genfromtxt(fname="lineintegralGauge_vof.csv",dtype=dtype,
                       delimiter=',',skip_header=1,names=names)
mpl.rcParams['figure.figsize'] ="10,"+str(N*2.5)
fig,ax = plt.subplots(N,1)
for j in range(N):
    ax[j].plot(gauges["t"],2.1-gauges[names[j+1]])
    ax[j].set_ylim([0,2.1])
```

```python
##velocity
N=5 #max 5
field='u'
dtype=tuple(['d']*(N+1))
names=['t']
for i in range(N):
    names.append(field+str(i+1))
#print dtype
#print names
gauges = np.genfromtxt(fname="pointGauge_u.csv",dtype=dtype,
                       delimiter=',',skip_header=1,names=names)
mpl.rcParams['figure.figsize'] = "10,"+str(N*2.5)
fig,ax = plt.subplots(N,1)
for j in range(N):
    ax[j].plot(gauges["t"],gauges[names[j+1]])
```
