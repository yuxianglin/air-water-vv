```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
#set_matplotlib_formats('svg')
import numpy as np
import matplotlib as mpl
from matplotlib  import pyplot as plt
```

```python
###pressure
N=13 # max 13
dtype=tuple(['d']*(N+1))
names=['t']
for i in range(N):
    field='p'+str(i+1)
    names.append(field)
#print dtype
#print names
gauges = np.genfromtxt(fname="pointGauge_pressure.csv",dtype=dtype,
                       delimiter=',',skip_header=1,names=names)
path='/workspace/Yuxiang/rice_tank/NHERIFuelTanks/data/inter/Solitary1/Trial04/'
time=np.loadtxt(path+'time.txt',comments='%')
time-=time[0]
time*=86400


expt=[time]
for j in range(N):
    name='press'+str(j+1)+'.txt'
    expt.append(np.loadtxt(path+name,comments='%')*6894.76)
    #print np.loadtxt(path+name,comments='%')*6894.76
mpl.rcParams['figure.figsize'] = "10,"+str(N*2.5)
fig,ax = plt.subplots(N,1)
for j in range(N):
    ax[j].plot(gauges["t"][3:],gauges[names[j+1]][3:],'-')
    ax[j].plot(expt[0],expt[j+1])
    #print gauges[names[j+1]][:20]
```

```python
### elevation
N=12 #max 12
field='h'
dtype=tuple(['d']*(N+1))
names=['t']
for i in range(N):
    names.append(field+str(i+1))
#print dtype
#print names
gauges = np.genfromtxt(fname="lineintegralGauge_vof.csv",dtype=dtype,
                       delimiter=',',skip_header=1,names=names)

path='/workspace/Yuxiang/rice_tank/NHERIFuelTanks/data/inter/Solitary1/Trial04/'
time=np.loadtxt(path+'time.txt',comments='%')
time-=time[0]
time*=86400
expt=[time]
for j in range(N):
    name='wg'+str(j+1)+'.txt'
    expt.append(np.loadtxt(path+name,comments='%'))
mpl.rcParams['figure.figsize'] ="10,"+str(N*2.5)
fig,ax = plt.subplots(N,1)
for j in range(N):
    ax[j].plot(gauges["t"],1.2-gauges[names[j+1]])
    ax[j].set_ylim([0,1.5])
    ax[j].plot(expt[0],expt[j+1]+0.5)
```
